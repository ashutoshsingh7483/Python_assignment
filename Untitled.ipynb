```python
import numpy as np
import pandas as pd
excelsheet=pd.ExcelFile('Python Assignment.xlsx')
sheet1=pd.read_excel(excelsheet,'(Input) User IDs')
sheet2=pd.read_excel(excelsheet,'(Input) Rigorbuilder RAW')
tab1=pd.DataFrame(sheet1,columns=['Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6']).loc[10:30]
tab2=pd.DataFrame(sheet2,columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6']).loc[7:29]



##############################output of sheet3########################


sheet3=pd.DataFrame(columns=['Team Rank','Thinking teams leaderboard','Average statements','average reasons','total'])
tab1['Unnamed: 5'] = tab1['Unnamed: 5'].replace(['Brandtech Lab'], 'BrandTech Lab')
team_list=tab1['Unnamed: 5']
team_list=set(team_list)
name_list=tab2['Unnamed: 3']
l=tab1.groupby('Unnamed: 5')['Unnamed: 4'].apply(list)
#print(l)
#print(name_list)
dic1={}
dic2={}
for i in name_list:
    dic1[i]=tab2.loc[tab2['Unnamed: 3']==i,'Unnamed: 5'].item()

for i in name_list:
    dic2[i]=tab2.loc[tab2['Unnamed: 3']==i,'Unnamed: 6'].item()
dic3={}
dic4={}
for i in team_list:
    lst=[]
    for j in l[i]:
        lst.append(dic1[j])
        avg=sum(lst)/len(lst)
    dic3[i]=avg
for i in team_list:
    lst=[]
    for j in l[i]:
        lst.append(dic2[j])
        avg=sum(lst)/len(lst)
    dic4[i]=avg   
l1=[]
l2=[]
l3=[]
l4=[]
for i in team_list:
    l1.append(i)
    l2.append(dic3[i])
    l3.append(dic4[i])
    tot=dic3[i]+dic4[i]
    l4.append(tot)
sheet3['Thinking teams leaderboard']=l1
sheet3['Average statements']=l2
sheet3['Average reasons']=l3
sheet3['total']=l4 
sheet3=sheet3.sort_values('total',ascending=False)
sheet3['Team Rank']=range(1,9)
del sheet3['total']
del sheet3['average reasons']
print('               OUTPUT OF SHEET 3             ')
print()
print(sheet3)

###########################output of sheet 4 in excel###################



sheet4=pd.DataFrame(columns=['Rank','name','uid','total_statement','total_reasons','total'])

sheet4['name']=tab2['Unnamed: 3']
sheet4['uid']=tab2['Unnamed: 4']
sheet4['total_statement']=tab2['Unnamed: 5']
sheet4['total_resons']=tab2['Unnamed: 6']
sheet4['total']=sheet4['total_statement']+sheet4['total_resons']
sheet4=sheet4.sort_values('total',ascending=False)
sheet4['Rank']=range(1,22)
del sheet4['total']
del sheet4['total_reasons']
print()
print()
print('               OUTPUT OF SHEET 4             ')
print()
print(sheet4)



        
        
    
    
```
